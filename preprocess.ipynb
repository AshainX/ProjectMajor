```python
''' 

CONTOUR IMAGE 

'''

import cv2
import os
import numpy as np

# Set the folder paths
input_folder = "/home/ashwathama/Documents/GitHub/2dshapedetection-major/testsegmentation"
output_folder = "/home/ashwathama/Documents/GitHub/2dshapedetection-major/testcontour"

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Function to preprocess and segment the image
def preprocess_and_segment_image(image):
    # Step 1: Apply Gaussian Blur to reduce noise
    blurred_img = cv2.GaussianBlur(image, (5, 5), 0)

    # Step 2: Morphological operation (Opening) to remove small noise
    kernel = np.ones((3, 3), np.uint8)  # Kernel for morphological operations
    morph_img = cv2.morphologyEx(blurred_img, cv2.MORPH_OPEN, kernel)

    # Step 3: Adaptive Thresholding for better segmentation
    threshold_img = cv2.adaptiveThreshold(morph_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY, 11, 2)

    # Step 4: Edge detection (Canny) to improve object boundaries
    edges_img = cv2.Canny(threshold_img, 50, 150)

    return edges_img

# Iterate through each file in the input folder
for filename in os.listdir(input_folder):
    file_path = os.path.join(input_folder, filename)

    # Ensure the file is an image (you can add more checks based on the image types)
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        # Read the binary image
        img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

        # Preprocess and segment the image
        segmented_img = preprocess_and_segment_image(img)

        # Find contours from the segmented edges
        contours, _ = cv2.findContours(segmented_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Create an empty image to draw contours
        contour_img = cv2.cvtColor(segmented_img, cv2.COLOR_GRAY2BGR)

        # Draw contours on the image
        cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 2)

        # Save the segmented image with contours drawn
        output_file_path = os.path.join(output_folder, f"segmented_{filename}")
        cv2.imwrite(output_file_path, contour_img)

        print(f"Processed {filename} and saved segmented image to {output_file_path}")

print("Image segmentation completed.")
```

```python
import cv2
import numpy as np
import os
from pathlib import Path

def preprocess_image(image):
    """
    Preprocess the image to prepare it for shape extraction.
    Converts to grayscale, applies thresholding and noise reduction.
    """
    # Convert to grayscale if the image is in color
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image.copy()
    
    # Apply Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    
    # Apply adaptive thresholding to handle different lighting conditions
    binary = cv2.adaptiveThreshold(
        blurred,
        255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY_INV,
        11,
        2
    )
    
    # Remove small noise using morphological operations
    kernel = np.ones((3,3), np.uint8)
    cleaned = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
    
    return cleaned

def extract_shapes(binary_image):
    """
    Extract geometric shapes from the binary image using contour detection.
    Filters out text and small artifacts.
    """
    # Find contours in the binary image
    contours, _ = cv2.findContours(
        binary_image,
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
    )
    
    # Create a blank image for drawing shapes
    shape_image = np.zeros_like(binary_image)
    
    # Filter and draw significant contours
    min_area = 100  # Adjust this threshold based on your images
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > min_area:
            # Draw the contour
            cv2.drawContours(shape_image, [contour], -1, 255, 1)
    
    return shape_image

def process_line_drawings(binary_image):
    """
    Special processing for line drawings to preserve thin lines.
    """
    # Apply Hough Line Transform
    edges = cv2.Canny(binary_image, 50, 150, apertureSize=3)
    lines = cv2.HoughLinesP(
        edges,
        1,
        np.pi/180,
        threshold=50,
        minLineLength=30,
        maxLineGap=10
    )
    
    # Create a blank image for drawing lines
    line_image = np.zeros_like(binary_image)
    
    # Draw detected lines
    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line[0]
            cv2.line(line_image, (x1, y1), (x2, y2), 255, 1)
    
    return line_image

def process_image(input_path, output_path):
    """
    Process a single image and save the result.
    """
    # Read the image
    image = cv2.imread(str(input_path))
    if image is None:
        print(f"Failed to read image: {input_path}")
        return
    
    # Preprocess the image
    binary = preprocess_image(image)
    
    # Extract shapes
    shapes = extract_shapes(binary)
    
    # Process line drawings
    lines = process_line_drawings(binary)
    
    # Combine shapes and lines
    result = cv2.bitwise_or(shapes, lines)
    
    # Save the result
    cv2.imwrite(str(output_path), result)

def process_folder(input_folder, output_folder):
    """
    Process all images in the input folder and save results to output folder.
    """
    # Create output folder if it doesn't exist
    output_path = Path(output_folder)
    output_path.mkdir(parents=True, exist_ok=True)
    
    # Process each image in the input folder
    input_path = Path(input_folder)
    for image_file in input_path.glob('*'):
        # Check if file is an image
        if image_file.suffix.lower() in ['.png', '.jpg', '.jpeg', '.tiff', '.bmp']:
            output_file = output_path / f"processed_{image_file.name}"
            print(f"Processing: {image_file}")
            process_image(image_file, output_file)

# Example usage
if __name__ == "__main__":
    input_folder = "/home/ashwathama/Documents/GitHub/2dshapedetection-major/TEST"
    output_folder = "/home/ashwathama/Documents/GitHub/2dshapedetection-major/CLAUDE_OUTPUT"
    process_folder(input_folder, output_folder)
```

```python
import cv2
import numpy as np
import os
from pathlib import Path

def preprocess_image(image):
    """
    Enhanced preprocessing specifically for circle detection.
    Uses multiple preprocessing steps to ensure better circle extraction.
    """
    # Convert to grayscale if the image is in color
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image.copy()
    
    # Increase contrast using CLAHE (Contrast Limited Adaptive Histogram Equalization)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    enhanced = clahe.apply(gray)
    
    # Apply bilateral filtering to reduce noise while preserving edges
    filtered = cv2.bilateralFilter(enhanced, 9, 75, 75)
    
    # Create two versions of binary images using different thresholding methods
    # This helps catch both bright and dark circles
    _, binary1 = cv2.threshold(filtered, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    binary2 = cv2.adaptiveThreshold(
        filtered,
        255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY,
        11,
        2
    )
    
    # Combine both binary images
    binary = cv2.bitwise_and(binary1, binary2)
    
    # Remove small noise using morphological operations
    kernel = np.ones((3,3), np.uint8)
    cleaned = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
    
    return cleaned

def detect_and_draw_circles(image, binary):
    """
    Specialized function for detecting and drawing circles.
    Uses both Hough Circle Transform and contour detection for better results.
    """
    # Create a blank image for drawing
    result = np.zeros_like(binary)
    height, width = binary.shape
    min_radius = min(height, width) // 10  # Minimum radius as a fraction of image size
    max_radius = min(height, width) // 2    # Maximum radius as a fraction of image size
    
    # Method 1: Hough Circle Detection
    circles = cv2.HoughCircles(
        image,
        cv2.HOUGH_GRADIENT,
        dp=1,
        minDist=50,
        param1=50,
        param2=30,
        minRadius=min_radius,
        maxRadius=max_radius
    )
    
    # Method 2: Contour Detection for circles
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Draw detected circles from Hough transform
    if circles is not None:
        circles = np.uint16(np.around(circles))
        for circle in circles[0, :]:
            center = (circle[0], circle[1])
            radius = circle[2]
            cv2.circle(result, center, radius, 255, 2)
    
    # Draw circles from contours
    for contour in contours:
        # Calculate contour properties
        area = cv2.contourArea(contour)
        perimeter = cv2.arcLength(contour, True)
        
        # Filter small contours
        if area < 100:  # Adjust this threshold based on your images
            continue
        
        # Check if contour is circular using circularity
        circularity = 4 * np.pi * area / (perimeter * perimeter)
        if circularity > 0.7:  # Threshold for circularity (1.0 is perfect circle)
            (x, y), radius = cv2.minEnclosingCircle(contour)
            center = (int(x), int(y))
            radius = int(radius)
            cv2.circle(result, center, radius, 255, 2)
    
    return result

def detect_arcs(binary):
    """
    Specialized function for detecting and drawing circular arcs.
    """
    result = np.zeros_like(binary)
    
    # Find contours
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    
    for contour in contours:
        # Fit an ellipse to the contour if possible
        if len(contour) >= 5:  # Need at least 5 points to fit an ellipse
            ellipse = cv2.fitEllipse(contour)
            # Draw the ellipse
            cv2.ellipse(result, ellipse, 255, 2)
    
    return result

def process_image(input_path, output_path):
    """
    Process a single image to extract circular shapes.
    """
    # Read the image
    image = cv2.imread(str(input_path))
    if image is None:
        print(f"Failed to read image: {input_path}")
        return
    
    # Preprocess the image
    preprocessed = preprocess_image(image)
    
    # Detect and draw circles
    circles = detect_and_draw_circles(preprocessed, preprocessed)
    
    # Detect and draw arcs
    arcs = detect_arcs(preprocessed)
    
    # Combine circles and arcs
    result = cv2.bitwise_or(circles, arcs)
    
    # Final cleaning using morphological operations
    kernel = np.ones((2,2), np.uint8)
    result = cv2.morphologyEx(result, cv2.MORPH_CLOSE, kernel)
    
    # Save the result
    cv2.imwrite(str(output_path), result)

def process_folder(input_folder, output_folder):
    """
    Process all images in the input folder and save results to output folder.
    """
    # Create output folder if it doesn't exist
    output_path = Path(output_folder)
    output_path.mkdir(parents=True, exist_ok=True)
    
    # Process each image in the input folder
    input_path = Path(input_folder)
    for image_file in input_path.glob('*'):
        # Check if file is an image
        if image_file.suffix.lower() in ['.png', '.jpg', '.jpeg', '.tiff', '.bmp']:
            output_file = output_path / f"processed_{image_file.name}"
            print(f"Processing: {image_file}")
            process_image(image_file, output_file)

# Example usage
if __name__ == "__main__":
    input_folder = "/home/ashwathama/Documents/GitHub/2dshapedetection-major/TEST"
    output_folder = "/home/ashwathama/Documents/GitHub/2dshapedetection-major/CLAUDE_OUTPUT2"
    process_folder(input_folder, output_folder)
```
